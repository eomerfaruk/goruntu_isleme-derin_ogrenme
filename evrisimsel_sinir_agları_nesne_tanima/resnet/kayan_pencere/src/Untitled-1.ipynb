```python
import os 
import cv2 
import numpy as np 
from matplotlib import pyplot as plt
```

```python
def sliding_window(img , step,ws):
    for y in range(0,img.shape[0]-ws[1],step):
        for x in range(0 , img.shape[1]-ws[0],step):

            yield(x,y,img[y:y+ws[1],x:x+ws[0]])
```
