```python
import os 
import cv2 
import numpy as np 

from collections import deque
```

```python
bufer_size = 16
pts = deque(maxlen = bufer_size)
```

```python
pts
```

```python
# mavi renk aralığı
blue_lower =(84,98,0)
blue_upper = (179,255,255)
```

```python
cap = cv2.VideoCapture(0) #cameradan görüntü almak için oluşturulan nesne
cap.set (3,960) #kamera genişlik
cap.set(4,450) #kamera yükseklik

while True:
    isopen , frame = cap.read()
    if isopen :
        blurred = cv2.GaussianBlur(frame ,(11,11) , 0) 
        hsv = cv2.cvtColor(blurred , cv2.COLOR_BGR2HSV)
        cv2.imshow("hsv img",hsv)
        mask = cv2.inRange(hsv , blue_lower , blue_upper)
        cv2.imshow("mask image",mask)

        # maskenin etrafında kalan görüntüleri sil . küçük noktalar silinir
        mask = cv2.erode(mask ,None, iterations=2)
        mask = cv2.dilate(mask , None , iterations=2)

        cv2.imshow("morfolojik mask",mask)

        contours , hierarchy = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL ,cv2.CHAIN_APPROX_SIMPLE) # sadece dıs contourları alır

        if len(contours) > 0: #contour varsa
            # en büyük contouru al
            c = max( contours , key = cv2.contourArea)
            #dikdörtgene cevir

            rect = cv2.minAreaRect(c)
            ((x,y),(w,h),rotation) = rect
            s = "x:{} , y:{} , w:{} , h:{} ,rotation:{}".format(np.round(x),np.round(y),np.round(w),np.round(h),np.round(rotation))
            print(s)
            # kutucuk (eğik) 
            box = cv2.boxPoints(rect)
            box = np.int64(box)
            
            # moment
            M = cv2.moments(c)
            center = (int(M["m10"] / M["m00"]),int(M["m01"] / M["m00"]))
            # contour cizdirme
            cv2.drawContours(frame ,[box],0,(0,255,255),2)

            # merkeze bir nokta cizme
            cv2.circle(frame ,center,5,(255,0,255),-1)      
            cv2.putText(frame ,s,(50,50),cv2.FONT_HERSHEY_SIMPLEX , 1,(0,0,0))

            #deque
            pts.appendleft(center)
            for i in range(1,len(pts)):
                if pts[i-10 ] is None or pts[i] is None:
                    continue
                cv2.line(frame,pts[i-1] , pts[i] , (0,255,0),3)            
            cv2.imshow("orjinal tespit" ,frame)

        if cv2.waitKey(1) & 0xFF == ord("q"): #q tusuna basılınca döngüden cıkılır
            break
    else:
        print("goruntu yok")        

cap.release()
cv2.destroyAllWindows()        
```
