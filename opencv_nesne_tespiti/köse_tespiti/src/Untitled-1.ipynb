```python
# shi thomasi 
img = cv2.imread("sudoku.png",0)
```

```python
img = np.float32(img)
corners = cv2.goodFeaturesToTrack(img, 100 ,0.01 ,10 )
```

```python
corners.shape
```

```python
corners = np.int64(corners)
```

```python
corners
```

```python
for i in corners:
    print(i)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(img ,(x,y),3,(125,125,125),cv2.FILLED)
```
