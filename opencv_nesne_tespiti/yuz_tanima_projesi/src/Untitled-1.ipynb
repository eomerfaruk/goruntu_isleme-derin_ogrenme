```python
insan_img = cv2.imread("einstein.jpg",0)
```

```python
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default (1).xml")
```

```python
face_rect = face_cascade.detectMultiScale(insan_img)
```

```python
for (x,y,w,h) in face_rect:
    cv2.rectangle(insan_img ,(x,y),(x+w,y+h),(255,255,255),3)
```

```python
plt.imshow(insan_img)
```

```python
img = cv2.imread("barcelona.jpg",0)
```

```python
face_rect = face_cascade.detectMultiScale(img)
```

```python
for (x,y,w,h) in face_rect:
    cv2.rectangle(img,(x,y),(x+w,y+h) , (255,255,255),4)
```

```python
plt.imshow(img)
```

```python
cap = cv2.VideoCapture(0)

while 1:
    ret , frame = cap.read()

    if ret :
        
        face_rect = face_cascade.detectMultiScale(frame , minNeighbors=7)
        
        for (x,y,w,h) in face_rect :
            cv2.rectangle(frame , (x,y),(x+w,y+h),(255,255,255),2)
        cv2.imshow("face detect",frame)    

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()

```
