```python
img = cv2.imread("contour.jpg",0)
```

```python
plt.imshow(img)
```

```python
contours , hierarchy = cv2.findContours(img , cv2.RETR_CCOMP , cv2.CHAIN_APPROX_SIMPLE) # iç ve dış contourların ikside bulunur
```

```python
external_contour = np.zeros(img.shape)
```

```python
internal_contour = np.zeros(img.shape)
```

```python
for i in range(len(contours)):
    if hierarchy[0][i][3] == -1: #external contour ise 
        cv2.drawContours(external_contour ,contours , i ,255,-1)
    else:
        cv2.drawContours(internal_contour ,contours , i ,255,-1)    
```
