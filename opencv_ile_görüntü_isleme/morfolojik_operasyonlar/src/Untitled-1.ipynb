```python
img = cv2.imread("team.png",0)
```

```python
plt.imshow(img)
```

```python
kernel = np.ones((5,5) , dtype = np.uint8)
```

```python
kernel
```

```python
result = cv2.erode(img , kernel ,iterations=2) # görüntünün ön kısmını genişliğini azaltır
```

```python
plt.imshow(result)
```

```python
result = cv2.dilate(img ,kernel , iterations = 2)
```

```python
plt.imshow(result)
```

```python
whitenoise = np.random.randint(0, 2 , size = img .shape[:2])
whitenoise = whitenoise * 255

noise_img = whitenoise + img
```

```python
plt.imshow(noise_img)
```

```python
opening = cv2.morphologyEx(noise_img.astype(np.float32) ,cv2.MORPH_OPEN , kernel) # opening: erode -> dilate 
```

```python
plt.imshow(opening)
```

```python
closed = cv2.morphologyEx(noise_img.astype(np.float32)  ,cv2.MORPH_CLOSE ,kernel)
```

```python
plt.imshow(closed)
```

```python
gradient = cv2.morphologyEx(img ,cv2.MORPH_GRADIENT , kernel )
```

```python
plt.imshow(gradient)
```
