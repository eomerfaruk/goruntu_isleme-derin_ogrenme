```python
def gaus_noise(img):
    row , col,ch = img.shape
    mean = 0 
    var = 0.05
    sigma = var ** 0.5

    gauss = np.random.normal(mean,sigma,(row,col,ch))
    gauss = gauss.reshape(row,col,ch)
    noisy = img  + gauss
    return noisy
```

```python
img = cv2.imread("manzara_constrat.jpg") / 255 # görüntünormalize edildi  
```

```python
plt.imshow(img) 
```

```python
noise_img = gaus_noise(img)
```

```python
plt.imshow(noise_img)
```

```python
blured = cv2.GaussianBlur(noise_img ,(3,3),7)
```

```python
plt.imshow(blured)
```
